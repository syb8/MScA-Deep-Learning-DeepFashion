```python
from imageai.Detection import VideoObjectDetection
from imageai.Detection.Custom import CustomObjectDetection
from imageai.Detection.Custom import DetectionModelTrainer

trainer = DetectionModelTrainer()
trainer.setModelTypeAsYOLOv3()
trainer.setDataDirectory(data_directory="./drive/MyDrive/Project YOLO/")
metrics = trainer.evaluateModel(model_path="./drive/MyDrive/Project YOLO/models/", json_path="./drive/MyDrive/Project YOLO/json/detection_config.json", iou_threshold=0.5, object_threshold=0.3, nms_threshold=0.5)
```

```python
metrics
```

```python
detector = CustomObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath(metrics[-1]['model_file'])
# detector.setModelPath("./drive/MyDrive/Project YOLO/models/detection_model-ex-013--loss-0015.210.h5")
detector.setJsonPath("./drive/MyDrive/Project YOLO/json/detection_config.json")
detector.loadModel()
```

```python
detections = detector.detectObjectsFromImage(input_image="./drive/MyDrive/Project YOLO/mmm_test.jpg", output_image_path="./drive/MyDrive/Project YOLO/mmm_test-detected.jpg")
for detection in detections:
    print(detection["name"], " : ", detection["percentage_probability"], " : ", detection["box_points"])
```
